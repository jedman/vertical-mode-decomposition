```python
data= nc.Dataset('sgpoldwpg.nc')
z = getvar('z',data)
theta = getvar('theta',data)
w = getvar('w', data)
tabs = getvar('tabs',data)
time = getvar('time',data)

tindex = np.where((time>132) & (time<134))
theta_sub = theta[tindex,:][0]
tabs_sub = tabs[tindex,:][0]
w_sub = w[tindex,:][0]
N_sq = vmd.brunt_vaisala(profile(theta),z)
time_sub = time[tindex]
```

```python
zi_full = vmd.make_zi_full(z) #make z interface vector including boundaries

eigs, zp = vmd.get_vertical_modes(N_sq, zi_full, rigidlid = 16368)

speeds = vmd.get_phase_speeds(eigs)

coeffs = []
for i in range(0,w_sub.shape[0]):
     coeffs.append(vmd.get_projection_coeffs(w_sub[i,:], eigs, N_sq, zp))

nmodes = 9 # looks at the first 9 modes 
tcomp = np.zeros((nmodes,len(coeffs)))

for i, val in enumerate(coeffs):
   for j in range(0,nmodes):
       tcomp[j,i] = val[j]
```

```python
dt = time[1]-time[0] # 5 minutes (units of days)
```

```python
from scipy import signal
```

```python
f, pgram = signal.periodogram(tcomp[0,:], 1./(dt))
f2, pgram2 = signal.periodogram(tcomp[1,:], 1./(dt))
f3, pgram3 = signal.periodogram(tcomp[2,:], 1./(dt))
f4, pgram4 = signal.periodogram(tcomp[3,:], 1./(dt))
f5, pgram5 = signal.periodogram(tcomp[4,:], 1./(dt))
```

```python
freq=[]
for speed in speeds: 
    freq.append(86400./(2*np.pi*128000./speed)) 
```

```python
fig, ax = plt.subplots(1)
ax.plot(f,pgram, marker='o',lw=3, label = 'Mode 1' )
ax.plot(f2,pgram2, marker='v',lw=3,label = 'Mode 2' )
ax.plot(f3,pgram3, marker='x', lw=3,label = 'Mode 3' )
ax.plot(f4,pgram4, marker='^', lw=3, label = 'Mode 4')

ax.legend(fontsize=14)
ax.set_xlim(0,10)
ax.set_ylim(0,0.8)
ax.vlines(freq[0:20], 0,0.8, lw=2)
ax.tick_params(labelsize=14)
ax.set_xlabel('cycles per day', fontsize=14)
fig.savefig('power_spectra.pdf')
```

```python
normedeigs=[] 
for index,eig in enumerate(eigs): 
    normedeigs.append(vmd.normalize(eigs,N_sq, zp, index))
levs = len(normedeigs[0])

mode1 = np.zeros( (levs,len(tindex[0])))
mode2 = np.zeros( (levs,len(tindex[0])))
mode3 = np.zeros( (levs,len(tindex[0])))
mode4 = np.zeros( (levs,len(tindex[0])))
mode5 = np.zeros( (levs,len(tindex[0])))
mode6 = np.zeros( (levs,len(tindex[0])))
mode7 = np.zeros( (levs,len(tindex[0])))
mode8 = np.zeros( (levs,len(tindex[0])))

modes = (mode1, mode2, mode3, mode4, mode5, mode6, mode7, mode8)
for i in range(len(time_sub)):
    for mindex, m in enumerate(modes): 
        m[:,i] = tcomp[mindex,i]*normedeigs[mindex]

    
```
